#Matplotlib
Matplotlib provides an object-oriented API for embedding plots into applications using general-purpose GUI toolkits like Tkinter, wxPython, QT, or GTK+.

```python
import matplotlib.pyplot as plt 
import numpy as np
```

# Scatter Plot - matplotlib

```python
x=np.arange(start=0,stop=10)
y=np.arange(11,21)
```

```python
plt.scatter(x,y,c='g')

#below are optional
plt.xlabel('X axis')
plt.ylabel('Y axis')
plt.title('Graph in 2D')

plt.savefig('graph.png') #saves a graph in an image format
```

```python
y=x*x
y
```

# Plot with diffrent styling - matplotlib

```python
plt.plot(x,y,'r--',linewidth=1)
```

```python
plt.plot(x,y,'r*',linestyle='dashed', linewidth=1, markersize=12)
```

```python
plt.subplot(2,2,1) #(number_rows, number_cols, pos)
plt.plot(x,y,'r')
plt.subplot(2,2,2)
plt.plot(x,y,'r*--')
plt.subplot(2,2,3)
plt.plot(x,y,'ro')
```

# Sine Waves - mtatplotlib

```python
np.pi
```

```python
x=np.arange(start=0, stop=4*np.pi, step=0.1)
y=np.sin(x)

plt.plot(x,y)
```

```python
x=np.arange(0,5*np.pi,0.1)
y_sin=np.sin(x)
y_cos=np.cos(x)
y_tan=np.tan(x)

plt.subplot(2,2,1)
plt.plot(x,y_sin)

plt.subplot(2,2,2)
plt.plot(x,y_cos,'r--')

plt.subplot(2,2,3)
plt.plot(x,y_tan,'g')

```

# Correlation with Heatmap
Correlation heatmaps are a type of plot that visualize the strength of relationships between numerical variables. Correlation plots are used to understand which variables are related to each other and the strength of this relationship.

Bsically it finds as interrelationship between each attribute. This cannot work on categorical fetures.

The corelation values always ranges from -1 to +1. i.e. specifically called as pearson correlation.




```python
df.dtypes
```

**Note**: As we know correlation cannot determine realtionship beween categorical variables, here are sex,smoker,day,& time are ctegorical feture, therefore it can not be determined

# Continuous Plots
Below plots are used for plotting continuous(integer, flotting-point) variables

# Joint Plot
The joint plot is the concise way of understanding the relationship between two variables as well as the individual distribution of each variable.

The Joint plot is consists of 3 separate plots. In which one is the middle figure which is used to see the relationship between x and y. This area gives us information about the joint distribution while the other two areas give us marginal distribution for the x and y-axis.

```python
sns.jointplot(x='tip', y='total_bill', data=df, kind='hex')
```

**Note:** In the above plot, there are more number of pepole who gave tip around 2,4 cent and their bill amount is between 10-20 cent.

```python
sns.jointplot(x='tip', y='total_bill', data=df, kind='reg')
```

# Pair Plot
The pairplot function creates a grid of Axes such that each variable in data will by shared in the y-axis across a single row and in the x-axis across a single column.

we can use pairplot when there are more than 2 independent fetures.

# Violin Plot
A violin plot is a hybrid of a box plot and a kernel density plot, which shows peaks in the data. It is used to visualize the distribution of numerical data. Unlike a box plot that can only show summary statistics, violin plots depict summary statistics and the density of each variable.

```python
sns.violinplot(x='tip', y='time', data=df, palette='rainbow')
```
